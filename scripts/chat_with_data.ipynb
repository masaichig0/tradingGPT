### Look into memory part. 

```python
# Use CSVLoader
from langchain.document_loaders.csv_loader import CSVLoader
loader = CSVLoader(file_path)
data = loader.load()
```

```python
data
```

```python
# Split text with ...
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=0)
```

```python
docs = splitter.split_documents(data[:500])
```

```python
docs
```

```python
# Embedding with `OpenAIEmbedding` model
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()
```

```python
# Use pinecone to do the same thing
import pinecone 
import os
from langchain.vectorstores import Pinecone
from tqdm.autonotebook import tqdm

# initialize pinecone
pinecone.init(
    api_key=os.environ.get("PINECONE_API_KEY"),  # find at app.pinecone.io
    environment= "us-east4-gcp"  # next to api key in console
)

index_name = "csvfile"
docsearch = Pinecone.from_documents(docs, embeddings, index_name=index_name, namespace="ifood_df")
```

```python
docsearch
```

```python
# Create prompt template with PromprTemplate
from langchain.prompts import PromptTemplate
from langchain.chains import ConversationalRetrievalChain
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain

chat = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo")
cd_template = """Given the following conversation and a follow up question, rephras the follow up question to be a standalone
question. 

Chat History:
{chat_history}
Follow Up Input: {question}
standalone question:"""
CONDENSE_PROMPT = PromptTemplate(template=cd_template, 
                                           input_variables=["chat_history", "question"])

qa_prompt = """You are AI Data Analyst and possesses a high level of ecpertise in analyzing and interpreting data to extract 
valuable insights and make data-driven decisions. You use various analytical tools and techniques to extract meaningful 
information from data. You also have passion to help business owner to give great result and you will do everything to make it
happen. I will give you a dataset, and ask yu a questions about what I want to know about the dataset. 

{context}
Question: {question}
Your Answer: 
"""

QA_PROMPT = PromptTemplate(template=qa_prompt, 
                                    input_variables=['context', 'question'])

# Create llm chain
chain = ConversationalRetrievalChain.from_llm(llm=chat, 
                                             retriever=docsearch.as_retriever(),
                                             condense_question_prompt=CONDENSE_PROMPT,
                                             qa_prompt=QA_PROMPT, 
                                             )
```

```python
# Chat with dataset
chat_history = []
question = "Can you tell me your insight about the data?"
result = chain({"question": question, "chat_history": chat_history})
```

## Try using Agent: 
first, use `VectorDBQA`

```python
from langchain import VectorDBQ
```

```python
# initialize Toolkit and Agent
from langchain.agents.agent_toolkits import (
    create_vectorstore_agent,
    VectorStoreToolkit,
    VectorStoreInfo,
)
vectorstore_info = VectorStoreInfo(
    name="ifood_df",
    description="marketing data from kaggle",
    vectorstore=docsearch
)
toolkit = VectorStoreToolkit(vectorstore_info=vectorstore_info)
agent_executor = create_vectorstore_agent(
    llm=chat,
    toolkit=toolkit,
    verbose=True
)
```

```python
agent_executor.run("How many rows does this dataset contains?")
```
